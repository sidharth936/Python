Statistical Feature Selection (Categorical Vs Continuous) using ANOVA test

Analysis of variance(ANOVA) is performed to check if there is any relationship between the given continuous and categorical variable

Assumption(H0): There is NO relation between the given variables (i.e. The average(mean) values of the numeric Predictor variable is same for all the groups in the categorical Target variable)
ANOVA Test result: Probability of H0 being true

```python
#Statistical Testing how much are the independent Variable coreelated with TargetVariable
def FunctionAnova(inptdata,TargetVariable,Predictor):
    from scipy.stats import f_oneway

    selectedPredictor=[]
    print('AnovaResult')

    for column in Predictor:
        CategoryGroupBy = inptdata.groupby(TargetVariable)[column].apply(list)
        AnovaResult = f_oneway(*CategoryGroupBy) 
        print(AnovaResult)
        if (AnovaResult[1] < 0.05):
            selectedPredictor.append(column)
            print(column,'is  correlated with',TargetVariable,'P Values is',AnovaResult[1])
        else:
            print(column,'is not correlated with',TargetVariable,'P Values is',AnovaResult[1])

    return(selectedPredictor)


        
        
    
```

```python
FunctionAnova(df_train, 'satisfaction',numeric)
```

Statistical Feature Selection (Categorical Vs Categorical) using Chi-Square Test
Chi-Square test is conducted to check the correlation between two categorical variables

Assumption(H0): The two columns are NOT related to each other
Result of Chi-Sq Test: The Probability of H0 being True

```python
def FunctionChiSq(inptdata,TargetVariable,Predictor):
    from scipy.stats import chi2_contingency
    
    selectedpredictor=[]
    for column in Predictor:
        CrossTab = pd.crosstab(index =inptdata[TargetVariable], columns =inptdata[column])
        chi2_contingency_result = chi2_contingency(CrossTab)
        if chi2_contingency_result[1] < 0.05:
            selectedpredictor.append(column)
            print(column, 'is correlated to',TargetVariable,'P Values is',chi2_contingency_result[1])
        else:
            print(column,'is not correlated to',TargetVariable,'P Value is',chi2_contingency_result[1])
    return(selectedpredictor)    
```

```python
FunctionChiSq(df_train, 'satisfaction', cat_cols)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
import xgboost as xgb

# assumes: X_train_scaled, X_test_scaled, y_train, y_test already exist

models = {
    'Logistic Regression': LogisticRegression(max_iter=2000, n_jobs=None),  # n_jobs for saga/liblinear not used
    'Linear SVM': LinearSVC(),
    'RBF SVM': SVC(kernel='rbf'),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'XGBoost': xgb.XGBClassifier(eval_metric='logloss', random_state=42, tree_method='hist'),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'KNN': KNeighborsClassifier()
}

results = []

for model_name, model in models.items():
    model.fit(X_train_scaled, y_train)

    y_train_pred = model.predict(X_train_scaled)
    y_test_pred  = model.predict(X_test_scaled)

    f1_train = f1_score(y_train, y_train_pred, average='weighted')
    f1_test  = f1_score(y_test,  y_test_pred,  average='weighted')

    results.append({
        'Model': model_name,
        'Parameters': model.get_params() if hasattr(model, 'get_params') else 'N/A',
        'F1 Train (weighted)': f1_train,
        'F1 Test (weighted)': f1_test,
    })

results_df = pd.DataFrame(results).sort_values('F1 Test (weighted)', ascending=False).reset_index(drop=True)
results_df
```
