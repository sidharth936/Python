```python
#Startup in which city have got funding 
CityWiseFunding = df.groupby('City  Location')['Amount in USD'].sum().sort_values(ascending = False).head(7)
```

```python
df['Year'] = df['Date dd/mm/yyyy'].dt.year
YearFunding = df.groupby('Year')['Amount in USD'].sum()
YearFunding
```

```python
#Different Year Funding
plt.figure(figsize =(8,6))
YearFunding.plot(kind = 'barh',x = YearFunding.index,  y= YearFunding.values, color = 'red')
plt.xlabel('Amount in USD')
plt.ylabel('Year')
plt.title('Year Wise funding')
plt.savefig('YearWiseStartUp.png', dpi=300)
plt.show()
```
